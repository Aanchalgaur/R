```r
library(extraTrees)
```

```r
## Classification with ExtraTrees (with test data)
  make.data <- function(n) {
    p <- 4   
    ##n- number of samples
    ##p- number of dimensions
    f <- function(x) (x[,1]>0.5) + (x[,2]>0.6) + (x[,3]>0.4)

    #x = input values
    #y= output values
    x <- matrix(runif(n*p), n, p)
    y <- as.factor(f(x))
    return(list(x=x, y=y))
  }
```

```r
train <- make.data(800)
  test  <- make.data(500)
  et    <- extraTrees(train$x, train$y)
  yhat  <- predict(et, test$x)
```
